```python
# Natural Language Toolkit — Python library used for working with text data (like cleaning, tokenizing, analyzing language, etc.)
import nltk
# nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
import re
stop_words = set(stopwords.words('english'))

def clean_text(text):
    if pd.isnull(text) or text == 'Unknown':
        return ""
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = [w for w in text.split() if w not in stop_words and len(w) > 2]
    return " ".join(words)

text_cols = ['CUSTOMER_VERBATIM', 'CAUSAL_VERBATIM', 'CORRECTION_VERBATIM']
for col in text_cols:
    df_subset[col + '_clean'] = df_subset[col].apply(clean_text)

df_subset['combined_text'] = (
    df_subset['CUSTOMER_VERBATIM_clean'] + ' ' +
    df_subset['CAUSAL_VERBATIM_clean']   + ' ' +
    df_subset['CORRECTION_VERBATIM_clean']
)

# Fix nulls in clean columns (caused by '#' only records)
df_subset['CUSTOMER_VERBATIM_clean'] = df_subset['CUSTOMER_VERBATIM_clean'].fillna(
    df_subset['CUSTOMER_VERBATIM'].astype(str).str.lower().str.replace(r'[^a-z\s]', '', regex=True).str.strip()
)
df_subset['CAUSAL_VERBATIM_clean'] = df_subset['CAUSAL_VERBATIM_clean'].fillna(
    df_subset['CAUSAL_VERBATIM'].astype(str).str.lower().str.replace(r'[^a-z\s]', '', regex=True).str.strip()
)

print("Text cleaning complete.")
print(f"Nulls in CUSTOMER_VERBATIM_clean : {df_subset['CUSTOMER_VERBATIM_clean'].isna().sum()}")
print(f"Nulls in CAUSAL_VERBATIM_clean   : {df_subset['CAUSAL_VERBATIM_clean'].isna().sum()}")


# ============================================================
# CELL 32 — ENTITY EXTRACTION (Unstructured → Structured)
# Brief format: Trigger | Failure Component | Failure Condition
#               | Additional Context | Fix Component | Fix Condition
# ============================================================

# --- Component keyword map (covers typos + multilingual) ---
components_map = {
    'radio':'Radio',        'raido':'Radio',          'radop':'Radio',
    'display':'Display',    'screen':'Screen',
    'camera':'Camera',      'module':'Module',
    'antenna':'Antenna',    'gps ant':'GPS Antenna',
    'amplifier':'Amplifier','tcicm':'TCICM',
    'bcm':'BCM',            'vcim':'VCIM',
    'usb':'USB',            'onstar':'OnStar',
    'navigation':'Navigation', 'sd card':'SD Card',
    'bluetooth':'Bluetooth','ethernet':'Ethernet Port',
    'steering':'Steering Module', 'speaker':'Speaker',
    'headunit':'Head Unit', 'head unit':'Head Unit',
    'connector':'Connector','harness':'Wiring Harness',
    'audio':'Audio Unit',   'switch':'Switch',
    'battery':'Battery',    'circuit':'Circuit',
    'port':'Port',
}

def extract_components(text, causal_cd=''):
    t = str(text).lower()
    found = []
    for kw, label in components_map.items():
        if kw in t and label not in found:
            found.append(label)
    # Fallback: derive from CAUSAL_CD_DESC if text has no component keywords
    if not found:
        cd = str(causal_cd).lower()
        if 'shorted' in cd:         found.append('Internal Component (Shorted)')
        elif 'communication' in cd: found.append('Communication Module')
        elif 'registers' in cd:     found.append('Control Module')
        elif 'broken' in cd:        found.append('Component (Broken)')
        elif 'poor contact' in cd:  found.append('Connector/Contact')
        elif 'discharge' in cd:     found.append('Battery')
        elif 'bulletin' in cd:      found.append('Software/Firmware')
        else:                       found.append('Infotainment System')
    return found[:3]

# --- Fix Condition map (English + French + Spanish + typos) ---
fix_cond_map = {
    'replaced':'Replaced',          'replace':'Replaced',
    'relac':'Replaced',             'r&r':'Replaced',
    'exchange':'Replaced',          'changed out':'Replaced',
    'repl ':'Replaced',             ' rr ':'Replaced',
    'r r ':'Replaced',              'rr radio':'Replaced',
    'programmed':'Programmed',      'program':'Programmed',
    'set up sps':'Programmed',      ' sps':'Programmed',
    'warr code':'Programmed',
    'updated':'Updated',            'update':'Updated',
    'usb file':'Updated',
    'reprogrammed':'Reprogrammed',  'reset':'Reset',
    'logic lock':'Reset',
    'flashed':'Flashed',            'installed':'Installed',
    'repaired':'Repaired',          'calibrat':'Calibrated',
    'remplacer':'Replaced',         'remplac':'Replaced',   # French
    'reemplaz':'Replaced',                                   # Spanish
    'vérifier':'Verified OK',       'verifier':'Verified OK',
    'verified':'Verified OK',       'verif':'Verified OK',
    'checks okay':'Verified OK',    'navigation is working':'Verified OK',
    'cleared':'Codes Cleared',      'relearn':'Relearn Performed',
    'not duplicated':'Not Duplicated',
    'no vehicle concern':'Not Duplicated',
    'unable to dup':'Not Duplicated',
    'no trouble':'Not Duplicated',  'no codes':'No Fault Found',
    'back order':'On Backorder',    'backorder':'On Backorder',
    'special order':'On Backorder', 'needs amplifier':'Needs Replacement',
    'diagnos':'Diagnosed',          'diag':'Diagnosed',
    'powers and grounds':'Diagnosed',
    'tested':'Tested',              'checked':'Inspected',
    'inspect':'Inspected',          'recheck':'Inspected',
    'ota':'OTA Update',
    'lube':'Cleaned/Lubed',         'clean':'Cleaned/Lubed',
    'internal failure':'Fault Identified',
    'internal short':'Fault Identified',
    'found internal':'Fault Identified',
    'short in radio':'Fault Identified',
    'no further':'No Further Action',
    'working on':'Under Investigation',
    'technical assistance':'TAC Case Opened',
    'claim code':'Warranty Claimed',
    'rental':'Rental Provided',
    'complete':'Completed',
}

def extract_fix_condition(text):
    t = str(text).lower()
    found = [v for k, v in fix_cond_map.items() if k in t]
    found = list(dict.fromkeys(found))[:2]
    return str(found) if found else 'Not Documented'

# --- Trigger map ---
trigger_map = {
    'reversing':'When Reversing',  'reverse':'When Reversing',
    'backup':'When Reversing',     'driving':'While Driving',
    'ignition':'On Ignition',      'startup':'On Startup',
    'wifi':'On WiFi Connect',      'bluetooth':'On Bluetooth',
    'volume':'Volume Adjustment',
}

def extract_trigger(text):
    t = str(text).lower()
    for kw, label in trigger_map.items():
        if kw in t:
            return label
    return 'No Additional Functionality'

# Apply all extractors
df_subset['Trigger']           = df_subset['CUSTOMER_VERBATIM'].apply(extract_trigger)
df_subset['Failure_Component'] = df_subset.apply(
    lambda r: extract_components(r['CAUSAL_VERBATIM'], r['CAUSAL_CD_DESC']), axis=1)
df_subset['Failure_Condition'] = df_subset['CAUSAL_CD_DESC']   # already structured
df_subset['Fix_Component']     = df_subset.apply(
    lambda r: extract_components(r['CORRECTION_VERBATIM'], r['CAUSAL_CD_DESC']), axis=1)
df_subset['Fix_Condition']     = df_subset['CORRECTION_VERBATIM'].apply(extract_fix_condition)

print("Entity extraction complete.")
print(f"Failure_Component unknowns  : {df_subset['Failure_Component'].apply(lambda x: len(x)==0).sum()}")
print(f"Fix_Component unknowns      : {df_subset['Fix_Component'].apply(lambda x: len(x)==0).sum()}")
print(f"Fix_Condition Not Documented: {(df_subset['Fix_Condition'] == 'Not Documented').sum()}")
print(f"\nTrigger distribution:")
print(df_subset['Trigger'].value_counts())
```

```python
# 2b. Issue Type Categorization
# Approach: Use CAUSAL_CD_DESC (structured column) as primary classifier
# + COMPLAINT_CD_DESC for specific sub-types.
# This avoids keyword overlap errors from raw free-text matching.
```

```python
print("Unique CAUSAL_CD_DESC values:")
print(df_subset['CAUSAL_CD_DESC'].unique())
print("\nUnique COMPLAINT_CD_DESC values:")
print(df_subset['COMPLAINT_CD_DESC'].unique())
```

```python
def categorize_issue(causal_cd, complaint_cd, verbatim_clean):
    causal    = str(causal_cd).lower()
    complaint = str(complaint_cd).lower()
    text      = str(verbatim_clean).lower()

    # Step 1: Specific sub-types from COMPLAINT_CD_DESC
    if 'navigation' in complaint and 'navigation' in text:
        return 'Navigation / GPS Issue'
    elif 'noise' in complaint or 'abnormal noise' in complaint:
        return 'Noise / Rattle Issue'
    elif 'video' in complaint or (
         'camera' in text and ('rear' in text or 'backup' in text)):
        return 'Rear Camera / Video Issue'

    # Step 2: Failure mode from CAUSAL_CD_DESC
    if 'no/incorrect communication' in causal:
        return 'Communication / Module Fault'
    elif 'shorted' in causal:
        return 'Short Circuit / Electrical Fault'
    elif 'registers incorrectly' in causal:
        return 'Incorrect Operation / Software Fault'
    elif 'broken' in causal or 'damaged' in causal or 'cracked' in causal:
        return 'Physical Damage / Hardware Fault'
    elif 'worn' in causal or 'stripped' in causal or 'poor contact' in causal:
        return 'Wear / Connection Fault'
    elif 'discharge' in causal:
        return 'Battery / Power Fault'
    elif 'incorrect version' in causal:
        return 'Software / Version Issue'
    elif 'campaign' in causal or 'tech bulletin' in causal:
        return 'Known Issue / Tech Bulletin'
    else:
        return 'General Infotainment Issue'

df_subset['Issue_Type'] = df_subset.apply(
    lambda r: categorize_issue(
        r['CAUSAL_CD_DESC'],
        r['COMPLAINT_CD_DESC'],
        r['CAUSAL_VERBATIM_clean']
    ), axis=1
)

print("Issue Type distribution:")
print(df_subset['Issue_Type'].value_counts())
```

```python
# 2b. Clustering — K-Means on TF-IDF
# TF-IDF converts text into numerical vectors.
# K-Means groups similar records based on word patterns.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
```

```python
vectorizer = TfidfVectorizer(
    max_features=2000,
    stop_words='english',
    ngram_range=(1, 2)   # unigrams + bigrams
)

# Use clean_text column (re-cleaned version)
def clean_text_v2(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df_subset['clean_text'] = df_subset['combined_text'].apply(clean_text_v2)

X = vectorizer.fit_transform(df_subset['clean_text'])
feature_names = vectorizer.get_feature_names_out()

print(f"TF-IDF matrix shape: {X.shape}")
print("Vectorization complete.")
```

```python
sum_words  = X.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for idx, word in enumerate(feature_names)]
words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)

print("Top 20 TF-IDF terms across all documents:")
for word, score in words_freq[:20]:
    print(f"  {word}: {float(score):.2f}")
```

```python
scores = []
for k in range(2, 8):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)
    print(f"K={k}, Silhouette Score={score:.4f}")

optimal_k = 5
print(f"\nSelected optimal_k = {optimal_k}")
```

```python
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_subset['Cluster'] = kmeans.fit_predict(X)
print("K-Means clustering complete.")
print(f"Cluster distribution:\n{df_subset['Cluster'].value_counts().sort_index()}")
```

```python
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]

print("\nTop words per cluster:\n")
cluster_keywords = {}
for i in range(optimal_k):
    words = [feature_names[ind] for ind in order_centroids[i, :10]]
    cluster_keywords[i] = words
    print(f"Cluster {i}: {words}")
```

```python
def name_cluster(words):
    # Most specific checks first — camera MUST come before program/usb

    if any(w in words for w in ['card', 'navigation', 'nav', 'sd']):
        return "Navigation / SD Card Issue"

    elif sum(w in words for w in ['screen', 'display', 'blank']) >= 2:
        return "Display / Screen Issue"

    elif any(w in words for w in ['camera', 'rear', 'coax', 'camera inop']):
        return "Rear Camera / Coax Issue"

    elif any(w in words for w in ['tac', 'update', 'usb', 'program', 'software', 'xm']):
        return "Software / Programming Issue"

    elif any(w in words for w in ['programmed', 'communication', 'control', 'ground', 'module']):
        return "Control Module / Electrical Issue"

    elif any(w in words for w in ['fault', 'internal fault', 'scan', 'internal', 'replace radio']):
        return "Internal Fault / Hardware Issue"

    else:
        return "General Infotainment Issue"


cluster_names = {}
for i in range(optimal_k):
    cluster_names[i] = name_cluster(cluster_keywords[i])

df_subset['Cluster_Name']  = df_subset['Cluster'].map(cluster_names)
df_subset['Cluster_Label'] = df_subset['Cluster'].astype(str) + " - " + df_subset['Cluster_Name']

print("Cluster assignments:")
for i in range(optimal_k):
    print(f"\nCluster {i} → {cluster_names[i]}")
    print(f"  Top words: {cluster_keywords[i]}")

print("\nCluster size distribution:")
print(df_subset['Cluster_Name'].value_counts())

# Verify no duplicate cluster names
unique_names = df_subset['Cluster_Name'].nunique()
print(f"\nUnique cluster names: {unique_names}")
if unique_names < optimal_k:
    print("WARNING: duplicate cluster names exist — check name_cluster() logic")
else:
    print("All clusters have unique names")
```
